## Basic Features

```python
eng_data = sales.dropna()
eng_data = eng_data.reset_index(drop=True)
eng_data['연속편성'] = 1

# 동일한 상품이 연속으로 편성될 경우 연속편성 +n
for idx in range(1,len(eng_data)):
    if eng_data['상품명'][idx] == eng_data['상품명'][idx-1]:
        eng_data['연속편성'][idx] += eng_data['연속편성'][idx-1]
sales = pd.merge(sales, eng_data[['방송일시', '상품코드', '연속편성']], how='left', on=['방송일시', '상품코드'])
# 동일 시간 같은 상품 노출시간, 연속편성 ffill로 채우기
sales = sales.fillna(method='ffill')
```

```python
evals['취급액'] = evals['취급액'].fillna(0) # 취급액은 모르는 값 
eng_data = evals.dropna()
eng_data = eng_data.reset_index(drop=True)
eng_data['연속편성'] = 1

# 동일한 상품이 연속으로 편성될 경우 연속편성 +n
for idx in range(1,len(eng_data)):
    if eng_data['상품명'][idx] == eng_data['상품명'][idx-1]:
        eng_data['연속편성'][idx] += eng_data['연속편성'][idx-1]
evals = pd.merge(evals, eng_data[['방송일시', '상품코드', '연속편성']], how='left', on=['방송일시', '상품코드'])
# 동일 시간 같은 상품 노출시간, 연속편성 ffill로 채우기
evals = evals.fillna(method='ffill')
```

```python
# NaN Preprocess
#sales['노출(분)'] = sales['노출(분)'].map(lambda x: round(x,-1))

#evals['노출(분)'] = evals['노출(분)'].map(lambda x: round(x,-1))

# Free product Preprocess 
sales = sales[sales['상품군']!='무형']

evals_free = evals[evals['상품군']=='무형']
evals = evals[evals['상품군']!='무형']


# Log y 
sales['log_y'] = np.log1p(sales['취급액'])

# 노출(분) 반올림 해주기
sales["노출(분)"]=sales["노출(분)"].apply(lambda x:round(x))
evals["노출(분)"]=evals["노출(분)"].apply(lambda x:round(x))

sales = sales[sales['판매단가']!=0]

# 판매단가 극단값 처리
quan_dict = {cat : 0 for cat in sales['상품군'].unique()}
for cat in sales['상품군'].unique():
    quan_dict[cat] = sales[sales['상품군']==cat]['판매단가'].quantile(0.90)
for idx in sales.index:
    if sales['판매단가'][idx] > quan_dict[sales['상품군'][idx]]:
        sales['판매단가'][idx] = quan_dict[sales['상품군'][idx]]
for idx in evals.index:
    if evals['판매단가'][idx] > quan_dict[evals['상품군'][idx]]:
        evals['판매단가'][idx] = quan_dict[evals['상품군'][idx]]
scaler = MinMaxScaler()
sales['판매단가'] = scaler.fit_transform(sales[['판매단가']])
evals['판매단가'] = scaler.transform(evals[['판매단가']])


print(sales.shape)
print(evals.shape)
```

## 변수 장착

```python
def feature_engineering(train, test) : 
    
    train['label'] = 'train'
    test['label'] = 'test'


    
    ### 기본 Feature Engineering #############################################################################
    
    ## times(노출 시간)의 횟수 (이전 누적시간은 조금 안맞다고 생각.)
    # why? -> 20분의 2번 누적은 40분. 그렇다면 30분 노출시간은 20분의 2번 노출보다 10분만큼 낮은 효과를 내는가?
    #         적절한 변수는 연속 노출 횟수라고 생각하여 변경 
    train = train.sort_values(['상품명','방송일시'])
    test = test.sort_values(['상품명','방송일시'])    
    
    
    
    ## born (기본 뼈대 작업)    
    y = train[['취급액','log_y']]
    data = pd.concat([train.drop(['취급액','log_y'], axis=1), test.drop(['취급액'],axis=1)], axis=0)
    
    ## feature 
    
    #1) times(노출 시간)
    data['month'] = data['방송일시'].map(lambda x: x.month)
    data['year'] = data['방송일시'].map(lambda x: x.year)
    data['day'] = data['방송일시'].map(lambda x: x.day)
    data['weekday'] = data['방송일시'].map(lambda x: x.weekday())
    data['minute'] = data['방송일시'].map(lambda x: x.minute)
    data['hour'] = data['방송일시'].map(lambda x: x.hour) 
    
    data[['month','weekday','hour']] = data[['month','weekday','hour']].astype('category')
    
    #2) 일시불=1, 무이자=2, 아무것도아니면=3 
    data["상품명2"]=data["상품명"].apply(lambda x:x.replace("(세일)","세일"))

    def sales_pil(x) : 
        if  ("일시불" in x or "일)" in x) :
            answer =  1
        elif ("무이자" in x or "무)" in x) :
            answer =  2
        else :
            answer = 0 
        return(answer)
    
    
    data['sales_way'] = data['상품명2'].map(lambda x: sales_pil(x)) 
    del data["상품명2"] 
    
    #3) 주차 데이터 
    data['weekofyear'] = data['방송일시'].apply(lambda x: x.isocalendar()[1])

    #4) 원플러스 데이터
    def oneplusone(x):
        if ("1+1" in x):
            return 1
        else : return 0
    data["oneplusone"] = data["상품명"].apply(lambda x:oneplusone(x))
    
    #5) 같이 광고되는 횟수
    together = data.groupby(['방송일시'])['상품명'].count().reset_index()
    together.columns = ['방송일시','together']
    data = pd.merge(data, together, how='inner', on=['방송일시'], right_index=True)
    
    #6) 아동이 포함돼있는가?
    def young_pil(x) :
        if  ("아동" in x or "남아" in x or "여아" in x) :
            answer =  1
        else :
            answer = 0
        return(answer)
    data['young'] = data['상품명'].map(lambda x: young_pil(x))
    
    #7) 성별은 어떠한가?
    def gender_pil(x) :
        if  ("남성" in x or "남아" in x or "(남)" in x) :
            answer =  1
        elif  ("여성" in x or "여아" in x or  "(여)" in x) :
            answer = 2
        else :
            answer = 0
        return(answer)
    data['gender'] = data['상품명'].map(lambda x: gender_pil(x))
    
    
    # return 
    train = data[data['label']=='train']
    train = pd.concat([train, y], axis=1)
    train = train.sort_index()
    test = data[data['label']=='test']
    test = test.sort_index()
    
    # 공휴일, 날씨 데이터 합치기 
    train = pd.merge(train, ts_dataset, how='left', on=['year','month','day']) 
    test = pd.merge(test, ts_dataset, how='left', on=['year','month','day'])
    
    train = pd.merge(train, whether, how='left', on=['year','month','day','hour'])
    test = pd.merge(test, whether, how='left', on=['year','month','day','hour'])
    
    # 미세먼지, 최고최저온도 합치기(일별)
    train = pd.merge(train, dust, how='left', on=['year','month','day']) 
    test = pd.merge(test, dust, how='left', on=['year','month','day'])
    
    # 개인카드사용현황
    train = pd.merge(train, card, how='left', on=['year','month']) 
    test = pd.merge(test, card, how='left', on=['year','month'])


    # 소비자물가지수
    train = pd.merge(train, consumer, how='left', on=['year','month','상품군']) 
    test = pd.merge(test, consumer, how='left', on=['year','month','상품군'])
    
    # 시청률 데이터 
    views['all_mean'] = views.iloc[:,1:366].apply('mean', axis=1) 
    views2 = views[['시간대','all_mean']]
    views2 = views2.drop(1440)

    views2['hour'] = views2['시간대'].map(lambda x:x[:2]).astype('int')
    views2['minute'] = views2['시간대'].map(lambda x:x[3:5]).astype('int')
    views2['time'] = views2['hour'] + views2['minute']/60

    train['time'] = train['hour'] + train['minute']/60
    test['time'] = test['hour'] + test['minute']/60
    train['views_time'] = 0.00000000000
    test['views_time'] = 0.00000000000

    for i in range(train.shape[0]) : 

        train['views_time'][i] = float(views2[(views2['time']>= train['time'][i]) & 
                                        (views2['time']<= train['time'][i]+train['노출(분)'][i]/60)]['all_mean'].mean())
        
    for i in range(test.shape[0]) : 
        test['views_time'][i] = float(views2[(views2['time']>= test['time'][i]) & 
                                        (views2['time']<= test['time'][i]+test['노출(분)'][i]/60)]['all_mean'].mean())
        
        
    
    return(train.drop(['label'],axis=1).sort_index(), test.drop(['label'],axis=1).sort_index())
```

## Word2vec 

### 상품명에 대한 차원벡터

```python
def word2vec_name(train, test, num_features) :  # sales, evals, number of demension

    def hash(astring):
        return ord(astring[0])
    
    text_data = pd.concat([train['상품명'], test['상품명']], axis=0).reset_index()
    text_data = text_data.drop(['index'],axis=1)
    
    corpus_path = text_data['상품명']
    sents = DoublespaceLineCorpus(corpus_path,iter_sent=True)

    word_extractor = WordExtractor(
        min_frequency=1, # hyper parameter !! 
        min_cohesion_forward=0.05,
        min_right_branching_entropy=0.
    )

    word_extractor.train(corpus_path)
    words = word_extractor.extract()

    cohesion_score = {word:score.cohesion_forward for word, score in words.items()}
    tokenizer = MaxScoreTokenizer(scores=cohesion_score)
    
    token_data = []
    def token_pil(text,i):

        for i in range(i):
            nouns = tokenizer.tokenize(text[i])
            token_data.append(nouns)

        return token_data


    token_data = token_pil(text_data['상품명'],len(text_data['상품명']))
    
    min_word_count = 1   # 단어에 대한 최소 빈도 수
    num_workers = 1      # 프로세스 개수, 2이상이면 seed를 고정해도 같은 값을 보장하지 않는다
    context = 10         # 컨텍스트 윈도우 크기
    downsampling = 1e-3   # 다운 샘플링 비율

    from gensim.models import word2vec

    print("Training model...")

    model = word2vec.Word2Vec(token_data,
                            workers=num_workers,
                            size=num_features,
                            min_count=min_word_count,
                            window=context,
                            sample=downsampling,
                             seed=200, hashfxn=hash)

    def get_features(words, model, num_features):
        # 출력 벡터 초기화
        feature_vector = np.zeros((num_features), dtype=np.float32)

        num_words = 0
        # 어휘 사전 준비
        index2word_set = set(model.wv.index2word)

        for w in words:
            if w in index2word_set:
                num_words += 1
                # 사전에 해당하는 단어에 대해 단어 벡터를 더함
                feature_vector = np.add(feature_vector, model[w])

        # 문장의 단어 수만큼 나누어 단어 벡터의 평균값을 문장 벡터로 함
        feature_vector = np.divide(feature_vector, num_words)
        return feature_vector

    def get_dataset(reviews, model, num_features):
        dataset = list()

        for s in reviews:
            dataset.append(get_features(s, model, num_features))

        reviewFeatureVecs = np.stack(dataset)
        return reviewFeatureVecs
    
    w2b= get_dataset(token_data,model,num_features)
    w2b_data = pd.DataFrame(w2b) 
    
    
    return(w2b_data)
```

```python
# 편성단위: 각각의 홈쇼핑 광고 편성
# 각 편성단위에 대해 고유번호 부여

opt['group'] = np.nan
k = 0

for i in range(opt.shape[0]):
    
    if not i : 
        opt.loc[i, 'group'] = k
        continue
    
    if opt.loc[i, 'together'] == opt.loc[i-1, 'together']:
        
        if (str(opt.loc[i, '마더코드'])[0:5] == str(opt.loc[i-1, '마더코드'])[0:5]) and (opt.loc[i, '방송일시'].hour - opt.loc[i-1, '방송일시'].hour <= 1):
            opt.loc[i, 'group'] = opt.loc[i-1, 'group']
            
        else:
            k += 1
            opt.loc[i, 'group'] = k
        
    else: 
        k += 1
        opt.loc[i, 'group'] = k

opt['group'].nunique()
```

```python
# 편성내용이 동일한 편성단위에 대해 동일한 고유번호 부여

group_case = defaultdict(list)
opt['code'] = np.nan

for i in range(opt.shape[0]):
    
    # 마더코드를 기준으로 번들 고유번호를 부여
    code = opt.loc[i, '마더코드']
    
    # group_case의 key값이 존재하지 않으면 편성코드를 새로운 key로 추가하고 key에 해당하는 value에 마더코드를 추가
    # 그 후, opt의 code 칼럼을 해당 편성코드(key)으로 채움
    if not group_case.keys():
        group_case[opt.loc[i, 'group']].append(code)
        opt.loc[i, 'code'] = opt.loc[i, 'group']
    
    # group_case에 존재하는 각각의 key에 대하여 value에 마더코드가 존재하는지 탐색
    for key in group_case.keys():
        
        # 특정 key 값에 마더코드가 존재하면 그 마더코드의 key 값으로 opt의 code 칼럼을 채움  
        if code in group_case[key]:
            opt.loc[i, 'code'] = key
    
    # 모든 key를 탐색했음에도 불구하고 마더코드를 발견하지 못한다면 편성코드를 새로운 key로 추가
    # 그 후, key에 해당하는 value에 마더코드를 추가
    # 마지막으로, opt의 code 칼럼을 해당 편성코드(key)로 채움
    if np.isnan(opt.loc[i, 'code']): 
        group_case[opt.loc[i, 'group']].append(code)
        opt.loc[i, 'code'] = opt.loc[i, 'group']
    
    
# 매핑 정확도 체크

for i in range(opt.shape[0]):
    
    try:
        # 자기자신 바로 위 row의 code값과 바로 아래 row의 code값을 before, after로 지정 
        before = opt.loc[i-1, 'code']
        after = opt.loc[i+1, 'code']
        
        # before과 after가 같다면 사이에 낀 자기자신의 code값을 before로 매핑
        if before == after:
            opt.loc[i, 'code'] = before
            
    except: continue    
     
    
opt['code'].nunique()
```

### 원 테스트 데이터로 계산한 황금시간대

```python
# 황금 시간대: 2020년 6월 모든 상품에 대한 weekday, hour 별 예상 취급액의 합계
# 예상 취급액이 높은 시간대부터 편성표 배치 시작

golden_time1 = test_table.groupby('방송일시').sum().reset_index()

golden_time1['weekday'] = golden_time1['방송일시'].dt.weekday
golden_time1['hour'] = golden_time1['방송일시'].dt.hour

golden_time1 = golden_time1.groupby(['weekday', 'hour']).mean().reset_index().sort_values(by='취급액', ascending=False).reset_index(drop=True)

print(golden_time1.shape)
golden_time1.head()
```

### 함수화

```python
# 주간 테이블을 생성하기 위해 원 테스트 데이터와 최적화용 데이터에서 각각 n주차 데이터만 추출

def weekly_chunk(nth_week):
    
    opt_week = opt_table.loc[opt_table['weekofyear'] == nth_week, :]
    test_week = test_table.loc[test_table['weekofyear'] == nth_week, :]
    
    return opt_week, test_week
```

```python
def whg_sum(nth_week, opt_week):
    
    # 각 번들의 요일, 시간별 예상 매출액
    # 황금 시간대를 기준으로 해당 시간대에 가장 높은 예상 취급액을 기록한 번들을 배치

    whg_sum = opt_week.groupby(['weekday', 'hour', 'group']).sum().reset_index().drop(columns='code')
    whg_sum = pd.merge(whg_sum, opt_week.drop_duplicates(['group'])[['group', '상품군', '상품명']], on='group', how='left')
    whg_sum = pd.merge(whg_sum, opt_week.drop_duplicates('상품명')[['상품명', 'code']], how='left')
    
    return whg_sum
```

```python
def get_timetable(whg_sum, golden_time):
    # 테이블 생성
    table = pd.DataFrame([[np.nan]*7]*24, dtype=object)
    table.columns.name = 'weekday'
    table.index.name = 'hour'
    # whg_sum: 각 번들의 요일, 시간별 예상 취급액
    # whg_sum을 참조 테이블에 복사
    ref_table = whg_sum.copy(deep=True)
    # 특정 요일에 배치된 번들 코드
    weekday_code = defaultdict(list)
    for i in range(ref_table['group'].nunique()):
        # golden_time: 2020년 6월 모든 상품에 대한 weekday, hour 별 예상 취급액의 합계
        # 황금 시간대의 weekday, hour
        weekday = golden_time.loc[i, 'weekday']
        hour = golden_time.loc[i, 'hour']
        # 황금시간대 weekday, hour에 대해 그 시간대에 가장 높은 예상 취급액을 기록한 상품 정보를 whg_sum 테이블에서 탐색
        search = ref_table.loc[(ref_table['weekday'] == weekday) & (ref_table['hour'] == hour)]
        try:
            # 방송 편성표의 셀이 채워지지 않은 채로 유지되는 동안 loop를 계속
            while isinstance(table.iloc[hour, weekday], float):
                # search에서 가장 높은 예상 취급액의 idx
                idx = search['취급액'].argmax()
                # idx로 상품 정보 탐색
                pgroup = search.iloc[idx].loc['상품군']
                group = search.iloc[idx].loc['group']
                code = search.iloc[idx].loc['code']
                pname = search.iloc[idx].loc['상품명']
                sales = search.iloc[idx].loc['취급액']
                # 만약 배치하려는 요일에 이미 동일한 번들 코드가 존재한다면 같은 요일에 2회 이상 배치할 수 없으므로,
                # search 테이블에서 해당 idx를 제외하고 상품을 배치하지 않음
                # 결과적으로 while loop를 돌면서 배치하려는 요일에 중복되는 번들 코드가 존재하지 않으면서,
                # 가장 높은 예상 취급액을 기록한 상품이 배치
                if code in weekday_code[weekday]:
                    search = search.drop(index=search.index[idx]).copy(deep=True)
                # 배치하려는 요일에 동일한 번들 코드가 존재하지 않는다면 곧바로 해당 weekday와 hour 위치에 상품을 배치
                # weekday_code 딕셔너리의 해당 요일 값에 번들 코드를 추가
                else:
                    weekday_code[weekday].append(code)
                    # 편성표의 각 셀을 (상품군, 편성그룹, 번들, 상품명, 예상 취급액)의 튜플로 채우기
                    table.iloc[hour, weekday] = (pgroup, group, code, pname, sales)
            # 배치가 끝나면 해당 시간대에는 더 이상 상품이 배치될 수 없으므로
            # 참조 테이블에서 해당 시간대 정보를 삭제
            ref_table = ref_table.loc[(ref_table['weekday'] != weekday) | (ref_table['hour'] != hour)]
            ref_table = ref_table.loc[ref_table['group'] != group]
        except:
            continue
    return table
```

```python
# 최적화 후 취급액 증가율

def opt_performance(table, test_week):

    before_opt = test_week['취급액'].sum()
    after_opt = table.fillna(' ').applymap(lambda x: x[-1] if x != ' ' else 0).sum().sum()
    
    print("최적화 이전 매출액 : " ,before_opt)
    print("최적화 이후 매출액 : " ,after_opt)
    
    opt_imp = ((after_opt - before_opt) / before_opt)*100
    
    return opt_imp
```
